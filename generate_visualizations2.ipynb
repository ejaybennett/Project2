```python
df5 = pd.DataFrame([], columns=['Medication', 'Frequency (Patient Level)'])
for medication in medication_list:
    total = 0
    for patient in medication_patient:
        if medication in medication_patient[patient]:
            total += 1
    df5.loc[len(df5.index)] = [medication,
                               total]

count_sum = sum(df5['Frequency (Patient Level)'])
df5['Percentage'] = (df5['Frequency (Patient Level)'] / count_sum) * 100

df5 = df5.sort_values(by=['Frequency (Patient Level)'], ascending=False)

df5.style.hide_index()
```
